```python
import seaborn as sns
import pandas as pd
```

```python
wine = pd.read_csv("Red.csv")
wine
```

```python
columns = ["Name", "Country", "Region", "Winery", "Rating", "NumberOfRatings", "Price", "Year"]
wine1 = wine[columns]
wine1
```

```python
diccio_columnas = {"Name" : "Nombre", "Country" : "País", "Region" : "Región", "Winery" : "Bodega", "Rating" :"Calificación", "NumberOfRatings" : "NºCal", "Price" : "Precio", "Year" : "Año"}
```

```python
wine1 = wine1.rename(columns = diccio_columnas)
wine1
```

```python
wine1.dropna( axis = 0, inplace = True)
wine1
```

```python
filtrado = wine1[(wine1["País"] == "Italy") | (wine1["País"] == "Spain")]
filtrado
```

```python
filtrado_wine = filtrado[filtrado["Calificación"] >= 4.5]
filtrado_wine.head(60)
```

```python
filtrado_wine.value_counts(["Calificación", "País"]).sort_index()
```

```python
filtrado2 = wine1[(wine1["País"] == "Spain")]
filtrado2.head(60)
```

```python
filtrado3 = filtrado2[filtrado2["Calificación"] >= 4.2]
filtrado3.head(50)
```

```python
filtrado3.value_counts(["Calificación", "Año"])
```

```python
filtrado3.value_counts(["Calificación", "Año"]).sort_index()
filtrado_ord = filtrado3.value_counts(["Calificación", "Año"]).sort_index()
filtrado_ord
```

```python
sns.jointplot(x=filtrado3.Calificación, y=filtrado3.Año)
```

```python
sns.histplot(x=filtrado3.Calificación, hue=filtrado3.Año, element="step")
```
